```python
df_reduced = df.drop(['id','date','sqft_living','waterfront', 'view', 'condition', 'grade',
                      'sqft_above', 'sqft_basement','yr_renovated','zipcode',
                      'lat', 'long', 'sqft_living15', 'sqft_lot15'],axis=1)
```

```python
df_reduced.columns
```

```python
X = df_reduced[['bedrooms', 'bathrooms', 'sqft_lot', 'floors', 'yr_built']]
y = df_reduced['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print('Intercept: ', lm.intercept_)
```

```python
pd.DataFrame(lm.coef_,X_train.columns,columns=['Coefficients'])
```

```python
predictions = lm.predict(X_test)
```

```python
sns.scatterplot(y_test, predictions)
```

```python
# Residual Histogram
sns.distplot((y_test-predictions),bins=30);
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
